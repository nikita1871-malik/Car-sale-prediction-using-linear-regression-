## step 4 creating model 

```python
import statsmodels.api as sm  

X = x_train[[
 #    'four_year_resale_value',
            'Price_in_thousands', 
             'Engine_size',
         # 'Horsepower',                  # all the columns will be markdown whose p value is more than 0.05
             'Wheelbase', 
           #   'Width',
           #  'Length', 
           #  'Curb weight',
       # 'Fuel_capacity',
             'Fuel_efficiency', 
           #   'Power_perf_factor',
       "Vehicle_type_Passenger"
        ]]

Y = y_train

X = sm.add_constant(X) # adding a constant for intercept as constant is the intercept 



model = sm.OLS(Y, X).fit()



print(model.summary())
```

```python
#Linear Regression 
#null Hypothesis : coeff of x1=coef of x2=coef of x3.......=0 (w1=w2=w3)"y does not depnd on any variable "
#alternate hypothesis = alteast one of coeff(wi) is not zero "y does depend on variable"
# t test 
#alpha 0.05
# any value with greater than 0.05 is useless to us as null will accept and null says that variable are not impo.
```
